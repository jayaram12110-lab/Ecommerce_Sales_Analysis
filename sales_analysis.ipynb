```python
import pandas as pd

df = pd.read_csv("Sample_ Superstore.csv")

df.head()
```

```python
df.columns
```

```python
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

print(category_sales)
```

```python
import matplotlib.pyplot as plt

category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
top_states = df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)

print(top_states)
```

```python
top_states.plot(kind="bar")

plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Sales")
plt.show()
```

## Top 10 States by Sales - Insight

California generated the highest sales among all states, followed by New York and Texas. These states contribute a significant portion of overall revenue, indicating strong customer demand and market presence. Businesses can focus on maintaining growth in these high-performing regions while exploring strategies to improve sales in lower-performing states.

```python
profit_state = df.groupby("State")["Profit"].sum().sort_values(ascending=False).head(10)

print(profit_state)
```

```python
import matplotlib.pyplot as plt

profit_state.plot(kind="bar")

plt.title("Top 10 States by Profit")
plt.xlabel("State")
plt.ylabel("Profit")
plt.show()
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
df["Order Date"].head(10)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")

print(df["Order Date"].dtype)
df["Order Date"].head()
```

```python
df["Order Date"] = pd.to_datetime(
    df["Order Date"],
    format="%m/%d/%Y"
)
```

```python
print(df["Order Date"].dtype)
print(df["Order Date"].head())
```

```python
monthly_sales = df.groupby(
    df["Order Date"].dt.to_period("M")
)["Sales"].sum()

monthly_sales.head()
```

```python
import matplotlib.pyplot as plt

monthly_sales.plot(figsize=(12,5))

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_products.plot(kind="bar")
plt.title("Top 10 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.xticks(rotation=75)
plt.show()
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
category_sales.plot(kind="pie", autopct="%1.1f%%")
plt.title("Category Wise Sales")
plt.ylabel("")
plt.show()
```

```python
region_profit = df.groupby("Region")["Profit"].sum()

plt.figure(figsize=(8,5))
region_profit.plot(kind="bar")
plt.title("Region Wise Profit")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(6,4))
sns.heatmap(
    df[["Sales","Quantity","Discount","Profit"]].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```
